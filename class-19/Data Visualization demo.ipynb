```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\Pavan\Downloads\fitness_data.csv")
df.head()
```

```python
plt.plot (df['steps'])
plt.title("raw steps data")
plt.ylabel("steps")
plt.xlabel("row of the data")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["date"],df["steps"])
plt.title("no of steps per day")
plt.xlabel("date")
plt.ylabel("no of steps")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['date'],df["resting_hr"],color="red",linewidth=.5)
plt.title("Resting heart rate over time",fontsize=12)
plt.xlabel("date")
plt.ylabel("resting hr(bpm)")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
##Scatter plt
plt.figure(figsize=(6,6))
plt.scatter(df['steps'],df["calories_burned"],alpha=0.6,color="slateblue")
plt.title("Calories Burned per steps",fontsize=15)
plt.xlabel("Steps",fontsize=12)
plt.ylabel("Calories Burned",fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
colors={"Low":"red","Moderate":"green","High":"blue"}

for level,subset in df.groupby("activity_level"):
    plt.scatter(subset['steps'],subset["calories_burned"],label= level,color=colors[level],alpha=0.6)
plt.title("steps vs calories  by activity level") 
plt.xlabel("steps") 
plt.ylabel("calories burned")  
plt.legend()
```

```python
##Histogram

plt.hist(df["steps"], bins=20, color="purple",edgecolor="black")
plt.title("Disribution of daily steps")
plt.ylabel("frequency")
```

```python
##subplots
fig, axes=plt.subplots(1,2)

axes[0].hist(df["sleep_hours"],bins=15,color= "teal",edgecolor="black")
axes[0].set_title("sleep hours Distribution")

axes[1].hist(df["resting_hr"],bins=20,color="red",edgecolor="black")
axes[1].set_title("Resting Heart rate")

plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(2,2)
#hist 1 (top left)
axes[0,0].hist(df["steps"],bins= 20,color="blue",edgecolor="black")
axes[0,0].set_title("daily steps")

#hist 2(top right)
axes[0,1].hist(df["calories_burned"],bins= 20,color="skyblue",edgecolor="black")
axes[0,1].set_title("Colories burned")

axes[1,0].hist(df["sleep_hours"],bins= 20,color="green",edgecolor="black")
axes[1,0].set_title("no of Hours")

axes[1,1].hist(df["resting_hr"],bins= 20,color="teal",edgecolor="black")
axes[1,1].set_title("resting heart rate")

plt.tight_layout()
plt.show()

```
